```python
dt = pd.get_dummies(data, columns=['employment_type', 'perform_cns_score_description'], drop_first=True)

# Convert average_acct_age and credit_history_length to numeric, handling non-numeric values
# Example: Convert to numeric while coercing errors to NaN
dt['average_acct_age'] = pd.to_numeric(data['average_acct_age'], errors='coerce')
dt['credit_history_length'] = pd.to_numeric(data['credit_history_length'], errors='coerce')

# Drop columns that cannot be converted or are not needed for correlation
dt.drop(columns=['uniqueid', 'date_of_birth', 'disbursaldate'], inplace=True)

# Calculate correlation with loan_default
correlation = dt.corr()['loan_default'].drop('loan_default')
```

```python
correlation
```

<h3>Positive Correlations:</h3>
<ul>
    <li><strong>Delinquent Accounts (Last 6 Months):</strong> A small increase in recent delinquencies is linked to a higher chance of default.</li>
    <li><strong>Number of Inquiries:</strong> More loan inquiries may slightly raise the risk of default.</li>
</ul>

<h3>Negative Correlations:</h3>
<ul>
    <li><strong>Perform CNS Score:</strong> A lower credit score is related to a higher risk of default. Better credit scores mean a lower chance of default.</li>
    <li><strong>Primary Active Accounts:</strong> Having more active loans might lower the risk of default.</li>
    <li><strong>Age:</strong> Older customers tend to have a lower probability of defaulting.</li>
</ul>

<h3>Low Impact Factors:</h3>
<ul>
    <li>Factors like loan amount, asset cost, and loan-to-value ratios show weak correlations with default risk, indicating they have little effect.</li>
</ul>

```python
# selecting relevant features based on the data dictionary
X = data[['disbursed_amount','asset_cost','ltv','age','perform_cns_score']]
y = data['loan_default']


# split the data into traning and test sets
X_train,X_test, y_train,y_test = train_test_split(X,y, test_size= 0.3, random_state = 42)

# Normalize data using normal scale
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)
```

```python
# Initialize and train the logistic model
model = LogisticRegression()
model.fit(X_train_scaled,y_train)

# make prediction on the test set
y_pred = model.predict(X_test_scaled)
```

```python
# Generate and display the confusion matrix 
conf_matrix = confusion_matrix(y_test, y_pred)
sns.heatmap(conf_matrix, annot= True, fmt='d',cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

```

```python
# Display the classification report
print(classification_report(y_test,y_pred))
```

<p>The model has accuracy of 79%</p>
